```python
import pandas as pd
import numpy as np
import random
```

```python
def arr_size(arr1,arr2,size):
    s=[]
    for i in range(0,int(len(arr2))+1,size):
        c=arr2[i:i+size]
        c=c.append(arr1)
        s.append(c)
    return s
```

```python
def arr_size1(arr1,arr2,size):
    s=[]
    for i in range(0,35401,size):
        c=arr2[i:i+size]
        c=c.append(arr1)
        s.append(c)
    return s
```

```python
def arr_size2(arr1,arr2,size):
    s=[]
    for i in range(0,40):
        c=arr2[0:size]
        c=c.append(arr1)
        s.append(c)
        arr2 = arr2.sample(frac=1.0)
    return s
```

```python
data= pd.read_csv(r'bulianxu40w.csv', sep=',')
data_= pd.read_csv(r'huapodian.csv',sep=',')
data = data.sample(frac=1.0)

res = arr_size2(data_,data,885)
for i in range(len(res)):
    data_df = pd.DataFrame(res[i])
    data_df.to_csv('./output2/res'+str(i+1)+'.csv',index=False)
```
